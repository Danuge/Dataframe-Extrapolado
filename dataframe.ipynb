```python
from random import choices, random
import pandas
import math

from collections import Counter
```

```python
def distribute(x, avg, std):
    return 0.5 * math.exp(-((x - avg)**2) / (2 * (std**2)))

def relacion_verb(x):
    return 1 + (0.0068743 * (x**1.35024)) - 0.516884

def relacion_num(x):
    return 1 + (0.00163259 * (x**1.97523)) - 0.511935

def relacion_raz(x):
    return 1 + (0.00840367 * (x**1.49395)) - 0.517565
```

```python
def hab(scores, avg, std, num=10**4):
    ch = []
    weights = []
    pc = []

    for i in range(1, scores+1):
        ch.append(i)

        weights.append(distribute(i, avg, std))

    pb = choices(ch, weights, k=num)

    for i in range(num):
        n = 0
        for j in range(num):
            if pb[j] < pb[i]:
                n +=1
        
        pc.append(max(1, 100*n // num))

    return pb, pc

def ipp(avg_a, std_a, avg_p, std_p, vb, vb_ch, rz, rz_ch, nm, nm_ch, num=10**4):
    
    act = []
    act_weights = []
    pr = []
    pr_weights = []

    samples = []

    for i in range(1, 100):
        act.append(i)
        pr.append(i)

        act_weights.append(distribute(i, avg_a, std_a))
        pr_weights.append(distribute(i, avg_p, std_p))

    act_samples = choices(act,act_weights,k=num)
    pr_samples = choices(pr,pr_weights,k=num)

    for i in range(num):
        if random() >= 1 - vb_ch:
            act_samples[i] = min(99, (act_samples[i] * relacion_verb(vb[i])))
        if random() >= 1 - vb_ch:
            pr_samples[i] = min(99, (pr_samples[i] * relacion_verb(vb[i])))

        if random() >= 1 - rz_ch:
            act_samples[i] = min(99, (act_samples[i] * relacion_raz(rz[i])))
        if random() >= 1 - rz_ch:
            pr_samples[i] = min(99, (pr_samples[i] * relacion_raz(rz[i])))

        if random() >= 1 - nm_ch:
            act_samples[i] = min(99, (act_samples[i] * relacion_num(nm[i])))
        if random() >= 1 - nm_ch:
            pr_samples[i] = min(99, (pr_samples[i] * relacion_num(nm[i])))

        samples.append(max(1, (act_samples[i] + pr_samples[i]) // 2))

    return samples
    
```

```python
verbal_pb, verbal_pc = hab(50, 29, 4)
razonamiento_pb, razonamiento_pc = hab(30, 22, 5)
num_pb, num_pc = hab(30, 18, 3)
```
